```python
from action_module import *
from generate_data import *
from draw_data import *
from statistics import variance 

np.random.seed(0)

ori_X, ori_y = generate_XOR_easy()
X_train = ori_X.tolist()
y_train = ori_y.squeeze().tolist()

Net = net(num_1 = 2, num_2 = 4, num_3 = 4, LR_ = 0.2)

EPOCH = 5000

loss_record = []
acc_record = []
var_record = []

for epoch in range(EPOCH):
    y_pred = [ ]
    total_loss = 0
    for X, y in zip(X_train, y_train):
        Net.forward(X)
        total_loss = total_loss + Net.loss(y)
        y_pred = y_pred + [int(Net.outlayer.y_pred>=0.5)]
        Net.cal_gradient()
        Net.update()
    
    print("====== Epoch %d/%d ======" % (epoch+1, EPOCH))
    
    buf = total_loss/len(X_train)
    loss_record = loss_record + [buf]
    print("loss: %.4f" % ( buf ) )
    
    buf = sum(np.array(y_pred)==ori_y.squeeze()) / len(X_train)
    acc_record = acc_record + [buf]
    print("acc: %.4f" % ( buf ))
    
    var_record = var_record + [variance(y_pred)]

```

```python
plt.figure(figsize=(10,5))
plt.plot(var_record)
plt.title("variance of y_pred")
plt.xlabel(" ")
plt.ylabel("loss ")
plt.show()
```
